```python

import sys  

reload(sys)  
sys.setdefaultencoding('utf8')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns


import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
cars = pd.read_csv("all.csv")
cars['year'] = cars['year'].convert_objects(convert_numeric=True) # convert year to a number
cars['age'] = 2017 - cars['year']

cars.maker = cars.maker.str.strip() # eliminate white spaces
cars.model = cars.model.str.strip() # eliminate white spaces

```

Best cars by maker
==

```python
makers = cars.groupby('maker')

maker_counts =  makers.count()
best_selling_cars = maker_counts[maker_counts.title>5].index


# Let's compute all the linear regressions for each car
lms = []
dlms = {}
for maker in best_selling_cars:
    lm = ols(formula="price ~ age", data=cars[cars.maker == maker]).fit()
#     m, b = np.polyfit()
    lms.append({ 'maker':maker,
                'b':lm.params[0],
                'm':lm.params[1],
                'count':maker_counts.loc[maker].year # this is the count
               })
    
    dlms[maker] = lm
    
#Convert the list to ta dataframe
lms = pd.DataFrame(lms)

# biggest= makers.filter(lambda x: len(x) > 500)


print lms.to_csv("lms_makers.csv")

```

Best cars by model
==

```python
models = cars.groupby(['maker', 'model'])

model_counts =  models.count()
best_selling_cars = model_counts[model_counts.title>5].index


# Let's compute all the linear regressions for each car
lms = []
dlms = {}
for (maker,model),values in models.groups.iteritems():
    # ignore cars with less than five listings
    if len(values)<5:
        continue
    lm = ols(formula="price ~ age", data=cars[cars.model == model]).fit()
#     m, b = np.polyfit()
    lms.append({ 'model':model,
                'b':lm.params[0],
                'm':lm.params[1],
                'maker':maker,
                'count':len(values) })
    
    dlms[model] = lm
    
#Convert the list to ta dataframe
lms = pd.DataFrame(lms)

# biggest= makers.filter(lambda x: len(x) > 500)



lms.to_csv("lms_models.csv")
lms
# sns.lmplot("age", "price", hue="maker",data= cars[cars.model.isin(best_selling_cars)])
```
